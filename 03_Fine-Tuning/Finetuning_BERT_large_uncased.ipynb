```python
#Create Directory for Annotation file 'dev_train.json'
!mkdir data
!mkdir data/squad20

# Store the file in 'data/squad20'
# Mount your Google Drive account
```

```python
#We Download here the Pre-trained model from HuggingFace.

reader = FARMReader(model_name_or_path="bert-large-uncased-whole-word-masking-finetuned-squad", use_gpu=True)
```

```python
train_data = "data/squad20"
reader.train(data_dir=train_data, train_filename="dev_train.json", use_gpu=True, n_epochs=2, save_dir="my_model_bert_large_uncased") #,num_processes=4)
```
